```python
import pandas as pd
df = pd.read_excel('autoreg_motor.xlsx')
df.head()
```

```python
subject_time = df[["subject#", "test_time", "motor_UPDRS", "yhat"]]
subject_time.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create the subplots
n_cols = 7
n_rows = 6
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols)
fig.set_size_inches(20, 10)
for i in range(42):
    data = subject_time[subject_time["subject#"] == i+1]
    axes[i//n_cols,i%n_cols].scatter(x = data["test_time"], y = data["motor_UPDRS"], color='blue')
    axes[i//n_cols,i%n_cols].scatter(data["test_time"], data["yhat"], color='black')
plt.tight_layout()
```
